```python
import matplotlib.pyplot as plt

class PaperTrader:
    def __init__(self, starting_cash=1000, transaction_cost=0.0075):
        self.starting_cash = starting_cash
        self.cash = starting_cash
        self.shares = 0
        self.transaction_cost = transaction_cost
        self.trade_log = []
        self.equity_curve = []

    def buy(self, quantity, price):
        cost = quantity * price
        fee = cost * self.transaction_cost
        if self.cash >= cost + fee:
            self.cash -= cost + fee
            self.shares += quantity
            self.trade_log.append(-cost)

    def sell(self, quantity, price):
        if self.shares >= quantity:
            value = quantity * price
            fee = value * self.transaction_cost
            self.cash += value - fee
            self.shares -= quantity
            self.trade_log.append(value)

    def portfolio_value(self, price):
        return self.cash + self.shares * price

    def update_equity(self, price):
        self.equity_curve.append(self.portfolio_value(price))
```

```python
# --- Simulation ---
trader = PaperTrader()
prices = [100, 102, 101, 105, 103]

for price in prices:
    trader.buy(1, price)
    trader.update_equity(price)

print("Cash:", round(trader.cash, 2))
print("Shares:", trader.shares)
print("Final Portfolio Value:", round(trader.portfolio_value(prices[-1]), 2))

plt.plot(trader.equity_curve, marker='o')
plt.title("Portfolio Equity Curve")
plt.xlabel("Trade Number")
plt.ylabel("Portfolio Value")
plt.grid(True)
plt.savefig("equity_curve.png")
plt.show()
```
